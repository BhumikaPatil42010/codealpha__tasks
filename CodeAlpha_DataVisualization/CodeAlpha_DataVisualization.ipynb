**STEP 3: Import & Load Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Zomato Restaurant names and Metadata.csv")
df.head()
```

**STEP 4 : Clean the Data**

```python
df.info()
df.isnull().sum()
```

```python
# Handle missing values:
df.dropna(inplace=True)
```

```python
# Remove duplicates:
df.drop_duplicates(inplace=True)
```

**STEP 5 : Start Creating Visualizations**

The analysis reveals that the restaurant industry is primarily driven by mid-range pricing and popular regional cuisines. The majority of restaurants operate in the affordable pricing segment, indicating a price-sensitive customer base.

Highly popular cuisines dominate the market, reflecting strong customer demand. However, these segments also show high competition levels, requiring differentiation strategies for new entrants.

**1. Bar Chart (Top Categories)**

```python
cuisine_count = df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(8,5))
plt.bar(cuisine_count.index, cuisine_count.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Cuisines by Restaurants")
plt.tight_layout()
plt.show()
```

* Graph Explanation â€“ Top 10 Cuisines by Restaurants

* This bar chart shows:

    X-axis â†’ Top 10 most common cuisines

    Y-axis â†’ Number of restaurants serving each cuisine

* Each bar represents how many restaurants offer that cuisine

ðŸ“Œ Insight:

* The bar chart highlights the top 10 cuisines served by restaurants. The leading cuisine shows significantly higher restaurant presence compared to others, indicating strong consumer demand. However, this also suggests higher competition within that category.

* Emerging cuisines with moderate representation may offer new business opportunities due to lower competition levels.

* Why This Graph Is Important

**2. Line Plot (Trend Analysis)**

```python
# Convert 'Cost' to numeric, handling commas
df['Cost_numeric'] = df['Cost'].str.replace(',', '').astype(int)

# Sort the DataFrame by Cost to visualize a trend
df_sorted_by_cost = df.sort_values(by='Cost_numeric').reset_index(drop=True)

plt.figure(figsize=(12, 6))
plt.plot(df_sorted_by_cost.index, df_sorted_by_cost['Cost_numeric'], marker='o', linestyle='-', color='skyblue')
plt.title("Restaurant Cost Trend (Sorted by Cost)")
plt.xlabel("Restaurant Index (Sorted by Cost)")
plt.ylabel("Average Cost for Two (in INR)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

* What the graph shows:

  X-axis â†’ Restaurant index (after sorting by cost)

  Y-axis â†’ Average cost for two people (in INR)

* The line increases from left to right

* Because you sorted by cost, the line shows a smooth upward trend.

  ðŸ‘‰ This graph does NOT show time-based trend.

  ðŸ‘‰ It shows distribution pattern of restaurant pricing.

ðŸ“Œ Insight:

The sorted line plot illustrates the distribution of restaurant pricing across the dataset. The majority of restaurants are concentrated within the low to mid-price range, indicating a price-sensitive market. Only a small proportion of restaurants fall into the high-cost category, suggesting that premium dining establishments represent a niche segment.

The gradual increase in the trend highlights a wide pricing spectrum, which can help businesses position themselves strategically within competitive pricing tiers.

* Business Interpretation

From a business perspective:

âœ” If opening new restaurant â†’ Target mid-range pricing

âœ” Premium segment â†’ Requires differentiation

âœ” Budget segment â†’ High competition

**3. Pie Chart (Distribution)**

```python
df['Collections'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))
plt.title("Restaurant Collections Distribution")
plt.ylabel('') # Hide the y-label for better aesthetics in pie chart
plt.show()
```

* What This Graph Shows

* Each slice represents a specific restaurant collection.

* The size of the slice represents its percentage share.

* Larger slice â†’ More restaurants in that category.

* Smaller slice â†’ Fewer restaurants in that category.

* This helps us understand category dominance.

ðŸ“Œ Insight:

The pie chart illustrates the distribution of restaurants across different collections. The largest segment represents the most popular category, indicating higher customer interest and stronger market presence in that segment.

Smaller segments reflect niche or specialized dining experiences. Businesses can leverage this information to identify high-demand segments or explore underrepresented categories for expansion opportunities.

**4. Histogram (Cost Distribution)**

```python

plt.figure(figsize=(10,6))

plt.hist(df['Cost_numeric'], bins=30)

plt.title("Distribution of Average Cost for Two")
plt.xlabel("Cost for Two (INR)")
plt.ylabel("Number of Restaurants")

plt.grid(True)
plt.tight_layout()
plt.show()

```

* What This Graph Shows:

  X-axis â†’ Cost range

  Y-axis â†’ Number of restaurants in each cost range

* Each bar â†’ How many restaurants fall into that price bracket

* The histogram reveals that the majority of restaurants fall within the low to mid-price range, indicating a price-sensitive customer base. Only a small proportion of restaurants operate in the premium segment, suggesting that luxury dining represents a niche market. This insight can help businesses determine optimal pricing strategies.

**5. Heatmap (Correlation)**

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

* It displays the correlation between numerical variables.

* Each cell shows correlation value between two variables.

* The correlation heatmap reveals relationships between numerical features in the dataset. A moderate positive correlation between cost and rating suggests that higher-priced restaurants may offer better quality experiences. Additionally, a positive relationship between votes and ratings indicates that popular restaurants tend to receive higher customer satisfaction scores.

Overall, no extremely strong correlations are observed, suggesting that multiple factors collectively influence restaurant performance.

* Heatmap helps in:

âœ” Feature selection for modeling
